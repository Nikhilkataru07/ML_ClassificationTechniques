```python
from sklearn.preprocessing import Imputer
imr=Imputer(missing_values=np.nan,strategy='most_frequent',axis=0)
df['MTT_50']=imr.fit_transform(df[['MTT_50']].values)
df['ETT_100']=imr.fit_transform(df[['ETT_100']].values)
df['ETP_100']=imr.fit_transform(df[['ETP_100']].values)
```

```python
print(df.isnull().values.any())
df.isnull().sum()
```

```python
#labelEncoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Termid']=le.fit_transform(df['Termid'].values)
df['Regd No']=le.fit_transform(df['Regd No'].values)
#df['Course']=le.fit_transform(df['Course'].values)
#df['Grade']=le.fit_transform(df['Grade'].values)
df['MHRDName']=le.fit_transform(df['MHRDName'].values)
df['ScholarType']=le.fit_transform(df['ScholarType'].values)
df['Direction']=le.fit_transform(df['Direction'].values)
#df['Gender']=le.fit_transform(df['Gender'].values)
df['Medium']=le.fit_transform(df['Medium'].values)
df['CourseType']=le.fit_transform(df['CourseType'].values)
df['ProgramType']=le.fit_transform(df['ProgramType'].values)
```
